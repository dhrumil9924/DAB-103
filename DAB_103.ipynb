## **Removing Unwanted Columns**

```python
data.drop(['Case Number', 'IUCR', 'Updated On', 'FBI Code', 'Beat'], inplace = True, axis =1)
```

```python
data.head()
```

## **Dividing time and date in different columns using to_datetime**

```python
Date = pd.to_datetime(data.Date, format = '%m/%d/%Y %I:%M:%S %p')
```

```python
data['Hour'] = Date.apply(lambda time: time.hour)
data['Month'] = Date.apply(lambda time: time.month)
data['Day of week'] = Date.apply(lambda time: time.dayofweek)
```

```python
data.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
data['Day of Week'] = data['Day of week'].map(dmap)
```

## **DATA VISUALIZATION IN GRAPHICAL FORM**

```python
sns.countplot(data=data, x = 'Arrest')
plt.show()
```

```python
sns.countplot(data=data, x = 'District')
plt.show()
```

```python
data.groupby('Primary Type').size().plot(kind='bar')
plt.title('Different type of Primary Type')
```

```python
data.groupby('Location Description').size().plot(kind='bar', figsize=(30,10))
plt.title('Number of crime by Location')
```

```python
sns.set(rc={'figure.figsize':(25,8.27)})
sns.countplot(x='Ward',data=data,hue='Arrest').set(title='Number of Arrest in Different Wards')
```

```python
pd.crosstab(data['Ward'],data['Arrest'])
```

```python
sns.countplot(x='Day of Week', data=data, hue='Primary Type').set(title='Type of different crime on different Days of Week')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```

```python
sns.countplot(x='Month', data=data).set(title='Number of crime on different Months')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```

```python
sns.countplot(x='Month', data=data, hue='Arrest').set(title='Number of people arrested in Months')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```

```python
sns.countplot(x='Arrest', data=data, hue='Primary Type').set(title="Type of different crime's criminal are arrested or not")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```
